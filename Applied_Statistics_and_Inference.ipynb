QUESTION-1: What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?

ANSWER:
  - Type I Error (α Error)

     A Type I error is a "false positive," where you reject the null hypothesis when it is actually true

     Concluding that there is an effect or difference when there isn't one
    
     EX- A medical test shows that a healthy person has a disease
     
     - Impact:

       Implementing a new, but ineffective, product, treatment, or policy. This can lead to wasted resources and a false belief in a breakthrough
  - Type II Error (β Error)

     A Type II error is a "false negative," where you fail to reject the null
     hypothesis when it is false
     
     EX-A quality control test fails to detect a flaw in a product, and the flawed products are released to customers.
     - Impact:

       Missing an opportunity to improve or innovate. It can mean an effective treatment isn't adopted or a product improvement is overlooked, potentially leading to lost sales or decreased profits

  - Impact on Decision-Making

     Type I error increases the risk of taking wrong actions based on false results.

     Type II error increases the risk of missing real problems or effects.

     Reducing one type of error often increases the other, so a balance must be maintained.


QUESTION 2: What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?

ANSWER:
- The p-value is the probability of obtaining results at least as extreme as the observed data, assuming the null hypothesis is true.

 - Interpreting the p-value in the context of the null hypothesis

   - Null Hypothesis (H₀): This is the default assumption of no effect, no difference, or no relationship. For example, (H₀) could state that a new drug has no effect on a disease.
   - P-value: This is the probability of observing your data (or more extreme data) if the null hypothesis were actually true.
   - Significance Level (α ): This is a pre-determined threshold, often set at (0.05) (or 5%). It is the probability of incorrectly rejecting the null hypothesis when it is actually true (a Type I error).

  - Comparison:

    - 1. If P-value ≤ α (significance level, e.g., 0.05):

         The result is statistically significant.

         We reject the null hypothesis (H₀).

         Meaning: The observed data is very unlikely under H₀.

     - 2. If P-value > α:

          The result is not statistically significant.

          We fail to reject the null hypothesis (H₀).

          Meaning: The observed data is likely under H₀, so no evidence against it.
  - Example- A drug trial has a p-value of (0.02) and the significance level is set at (0.05).Since (0.02 < 0.05),

     you reject the null hypothesis

     There is only a 2% chance of observing such data if H₀ were true.

QUESTION-4: What is a confidence interval, and how does the margin of error
           influence its width and interpretation

ANSWER:
- A confidence interval(CI) is a range of values, calculated from sample data, that is likely to contain an unknown population parameter. and The margin of error determines the width of this interval and represents the amount of random sampling error

  
  Example: A 95% CI of (40, 50) means we are 95% confident that the true value lies between 40 and 50.
- How the margin of error influences the interval Width:
    - The margin of error is added to and subtracted from the point estimate to create the interval (Point Estimate (±) Margin of Error). Therefore, a larger margin of error directly creates a wider interval.
    - Precision: A wider interval (due to a larger margin of error) indicates more uncertainty and a less precise estimate. A narrower interval (smaller margin of error) is more precise.
- Factors that increase the margin of error (and thus widen the interval):
    - Higher confidence level: To be more confident, you need a wider range to capture the true value.Lower sample size: A smaller sample has more uncertainty.
    - Higher sample variability: If the data points in the sample are very spread out, the estimate is less certain

- How to interpret a confidence interval
  
  The width of the confidence interval directly depends on the margin of error. A wide CI shows less precision, while a narrow CI shows more precision in estimating the population paramete

QUESTION-5: Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?

ANSWER:
- Purpose of ANOVA-Its purpose is to compare the variance between groups to the variance within groups to see if the variation between the groups is significantly larger than what would be expected by chance. It extends hypothesis testing beyond two groups

- Assumptions of ANOVA

  For the results of an ANOVA test to be valid, certain assumptions must be met:

  - Random sampling: Data must be randomly sampled, and observations must be independent.
  - Normal distribution: The dependent variable should be approximately normally distributed for each group.
  - Homogeneity of variances: The variance within each group should be roughly equal (homoscedasticity).
  - Continuous dependent variable: The dependent variable must be continuous and measured on an interval or ratio scale.
  - Independent groups: The independent variable should consist of two or more categorical, independent groups
- How it extends hypothesis testing to more than two groups
   - Single test for multiple comparisons: ANOVA allows you to test the null hypothesis that all group means are equal in a single test, whereas using multiple t-tests would be inefficient and increase the probability of a Type I error.
   - Comparison of variances: It works by comparing the variance between the groups (the variation in the means) to the variance within each group (the variation of individual data points around their group mean).
   - F-statistic: The result of this comparison is the F-statistic, or F-ratio. If the variation between groups is significantly larger than the variation within groups, the F-statistic will be large, indicating that there is a significant difference between at least some of the group means.
   - Rejection of the null hypothesis: If the F-statistic is large enough (based on an F-distribution), the null hypothesis that all means are equal is rejected, leading to the conclusion that at least one group mean is different from the others



```python
'''
QUESTION-6 Write a Python program to perform a one-sample Z-test and interpret
 the result for a given dataset.
'''
import numpy as np
from math import sqrt
from scipy.stats import norm

# Sample data
data = [52, 55, 50, 49, 53, 54, 51, 52, 56, 50]

# Known population mean (H0)
mu_0 = 50

# Known population standard deviation
sigma = 4

# Calculate sample mean
sample_mean = np.mean(data)

# Sample size
n = len(data)

# Calculate Z statistic
z_value = (sample_mean - mu_0) / (sigma / sqrt(n))

# Calculate p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_value)))

print("Sample Mean:", sample_mean)
print("Z-value:", z_value)
print("P-value:", p_value)

alpha = 0.05

# Interpretation
if p_value < alpha:
    print("Result: Reject the null hypothesis (H0).")
    print("There is significant evidence that the population mean is different from", mu_0)
else:
    print("Result: Fail to reject the null hypothesis (H0).")
    print("There is no significant evidence that the population mean differs from", mu_0)
'''
Explanation of the Program
1. Input
A sample dataset
Known population mean (H₀)
Known population standard deviation (required for Z-test)
2. Calculations
Computes sample mean
Computes Z-value using:
Computes p-value (two-tailed)
3. Interpretation
If p-value < α (0.05) → Reject H₀ (significant difference)
If p-value > α (0.05) → Fail to reject H₀ (no significant difference)
'''
```

```python
'''
QUESTION-8 Generate multiple samples from a non-normal distribution and
implement the Central Limit Theorem using Python
'''
'''The Central Limit Theorem (CLT) states that the distribution of sample means
of a large number of independent, identically distributed random variables
approaches a normal distribution, regardless of the shape of the original
distribution.
'''
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate many samples from a non-normal distribution (Exponential)
num_samples = 10000      # total number of samples
sample_size = 50         # each sample ka size

sample_means = []        # list to store means

for i in range(num_samples):
    sample = np.random.exponential(scale=1, size=sample_size)
    sample_means.append(np.mean(sample))

# Step 2: Plot histogram of sample means (CLT shows approx Normal)
plt.hist(sample_means, bins=40)
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.title("Central Limit Theorem Demo (Exponential → Normal)")
plt.show()
```

```python
'''
QUESTION-9 Write a Python function to calculate and visualize the confidence
 interval for a sample mean.
'''
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from scipy.stats import norm

def plot_confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)

    # Standard error
    se = sample_std / sqrt(n)

    # Critical value
    z = norm.ppf((1 + confidence) / 2)

    # Confidence interval
    lower = sample_mean - z * se
    upper = sample_mean + z * se

    # Plot
    plt.figure(figsize=(8,4))
    plt.axvline(sample_mean, label="Sample Mean")
    plt.axvspan(lower, upper, alpha=0.3)
    plt.title(f"Confidence Interval ({confidence*100:.0f}%)")
    plt.xlabel("Value")
    plt.legend()
    plt.show()

# Example data
data = np.random.randn(50) * 5 + 40  # random sample
plot_confidence_interval(data)
```

QUESTION-10: Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome.

ANSWER:
- The Chi-square goodness-of-fit test determines if an observed frequency distribution significantly differs from an expected distribution



```python
from scipy.stats import chisquare
import numpy as np

# Observed frequencies
observed_frequencies = np.array([22, 30, 23])

# Expected frequencies (e.g., based on a theoretical distribution or equal distribution)
expected_frequencies = np.array([25, 25, 25]) # Assuming equal distribution for example

# Perform the Chi-square goodness-of-fit test
chi2_statistic, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

print(f"Chi-square statistic: {chi2_statistic:.2f}")
print(f"P-value: {p_value:.3f}")

# Interpret the outcome
alpha = 0.05 # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The observed distribution is significantly different from the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed distribution does not significantly differ from the expected distribution.")
```

- Explanation of the Outcome:
   - Chi-square Statistic: This value quantifies the discrepancy between the observed and expected frequencies. A larger chi-square statistic indicates a greater difference between the distributions.
   - P-value: The p-value represents the probability of observing a chi-square statistic as extreme as, or more extreme than, the one calculated, assuming the null hypothesis is true (i.e., the observed and expected distributions are the same).
- Interpretation:
  - If p-value < alpha (significance level, commonly 0.05): The result is statistically significant. This means there is sufficient evidence to reject the null hypothesis, concluding that the observed distribution is significantly different from the expected distribution.
  - If p-value >= alpha: The result is not statistically significant. There is insufficient evidence to reject the null hypothesis, suggesting that the observed distribution does not significantly differ from the expected distribution. The differences observed could be due to random chance
