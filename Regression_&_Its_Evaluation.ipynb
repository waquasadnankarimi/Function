Question 1: What is Simple Linear Regression?

Answer:
- Simple Linear Regression is a statistical technique used to find the relationship between one independent variable and one dependent variable. It models this relationship using a straight line equation:
   y=mx+c


- where Y is the dependent variable, X is the independent variable, m is the   slope of the line, and c is the intercept.
It is mainly used for prediction and to understand how changes in the independent variable affect the dependent variable.

Question 2: What are the key assumptions of Simple Linear Regression?

Answer:
- The key assumptions of Simple Linear Regression are:

  - Linearity: There is a linear relationship between the independent variable (X) and the dependent variable (Y).

  - Independence: The observations are independent of each other.

  - Homoscedasticity: The variance of the errors is constant for all values of X.

  - Normality of Errors: The residuals (errors) are normally distributed.

  - No Multicollinearity: There is only one independent variable, so multicollinearity does not exist in Simple Linear Regression.

Question 3: What is heteroscedasticity, and why is it important to address in regression models?

Answer:
 - Definition: The non-constant variance of the residuals (errors) across different levels of the predictor variables, often appearing as a fan or cone shape in residual plots.
 - Opposite: Homoscedasticity (constant variance) is the desired condition in standard regression.
 -Example: As income rises, food expenditure variability increases (some spend a lot, others little), while for low-income earners, expenditure is consistently low.
**Why It's Important to Address**
 - Unreliable Standard Errors: OLS (Ordinary Least Squares) assumes homoscedasticity; heteroscedasticity breaks this, causing standard errors to be biased (usually underestimated).
 - Invalid Hypothesis Tests: Incorrect standard errors mean t-statistics and p-values are wrong, making you think a variable is significant when it isn't (Type I error) or vice-versa.
 - Misleading Confidence Intervals: Confidence intervals become too narrow or wide, leading to incorrect predictions and interpretations.
 - Inefficient Estimates: While coefficients remain unbiased, they are no longer the "Best Linear Unbiased Estimators" (BLUE)

 Question 4: What is Multiple Linear Regression?

 Answer:
 - Multiple Linear Regression is a statistical technique used to model the relationship between one dependent variable and two or more independent variables. It explains how the dependent variable changes when any one of the independent variables is varied, while the others are kept constant.

 The mathematical form of Multiple Linear Regression is:
 - Y=b0​+b1​X1​+b2​X2​+⋯+bn​Xn​
 - where Y is the dependent variable, X₁, X₂, …, Xₙ are independent variables, b₀ is the intercept, and b₁, b₂, …, bₙ are the regression coefficients.

 - Multiple Linear Regression is widely used for prediction, analysis, and understanding the combined effect of multiple factors on an outcome variable.

Question 5: What is polynomial regression, and how does it differ from linear
regression?

Answer:
- Polynomial regression models curved relationships by adding polynomial terms (x^2,x^3 etc.) to an equation, making it fit complex data, while linear regression fits straight lines for simple, linear trends, differing in flexibility, complexity, and ability to capture non-linear patterns. Linear models assume a straight-line relationship \(Y=\beta _{0}+\beta _{1}X\), whereas polynomial models can fit curves (\(Y=\beta _{0}+\beta _{1}X+\beta _{2}X^{2}+\dots +\beta _{n}X^{n}\)).

| Basis        | Linear Regression            | Polynomial Regression                           |
| ------------ | ---------------------------- | ----------------------------------------------- |
| Relationship | Linear (straight line)       | Non-linear (curve)                              |
| Equation     | (Y = \beta_0 + \beta_1 X)    | (Y = \beta_0 + \beta_1 X + \beta_2 X^2 + \dots) |
| Graph shape  | Straight line                | Curved line                                     |
| Complexity   | Simple                       | More complex                                    |
| Usage        | When data shows linear trend | When data shows curved trend                    |

- Conclusion

 - Linear regression is suitable when the relationship between variables is straight-line, whereas polynomial regression is used when the relationship is curved or non-linear, providing a better fit for complex data patterns.

Question 10: Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location.
However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust model.

Answer:
- As a data scientist addressing common challenges in a real estate price prediction model, I would follow specific steps to mitigate heteroscedasticity and multicollinearity, thereby ensuring a more robust and reliable model [1, 2]

**Addressing Heteroscedasticity**
- Heteroscedasticity occurs when the variance of the errors (residuals) is not constant across all levels of the independent variables. This leads to inefficient parameter estimates and unreliable hypothesis tests [1]
  - Transform the Dependent Variable: A common approach is to apply a logarithmic transformation to the house price data. This often stabilizes the variance and makes the data distribution more symmetric [1, 2].
  - Use Robust Standard Errors: I would employ the use of robust standard errors (also known as Huber-White standard errors). These standard errors provide consistent estimates even in the presence of heteroscedasticity, ensuring that confidence intervals and p-values are reliable [1].
  - Explore Weighted Least Squares (WLS): If the pattern of the variance is known (e.g., variance increases with house size), I would use WLS. This technique assigns different weights to observations based on their variance, giving less weight to observations with larger errors and more weight to those with smaller errors [1].
  - Visualize Residuals: Throughout the process, I would use residual plots (e.g., residuals versus fitted values) to visually confirm if the steps taken have successfully normalized the spread of the residuals [2]

Addressing Multicollinearity **bold text**
- Multicollinearity arises when independent variables in a regression model are correlated with each other. This makes it difficult to ascertain the individual effect of each predictor on the target variable and can lead to unstable and counterintuitive coefficient estimates [2, 1],
  - Calculate Variance Inflation Factor (VIF): I would first quantify the degree of multicollinearity by calculating the VIF for each independent variable. A VIF value above a certain threshold (commonly 5 or 10) indicates significant multicollinearity [1, 2]
  - Feature Selection:
     - Remove Highly Correlated Features: If two features (e.g., total square footage and number of bedrooms in small apartments) are highly correlated, I would remove one of them from the model [1, 2].
     - Combine Features: Alternatively, I could combine correlated features into a single composite feature (e.g., a "property size index") [2].

 - Use Regularization Techniques: Techniques like Ridge Regression are specifically designed to handle multicollinearity. Ridge regression adds a penalty term to the loss function, which shrinks the regression coefficients and mitigates the impact of highly correlated features on the model's stability [1, 2].
 - Principal Component Analysis (PCA): As a more advanced approach, I might use PCA to transform the original correlated features into a new set of uncorrelated components (principal components) [1]. I would then use a selection of these components as predictors in the regression model

```python
'''
Question 6: Implement a Python program to fit a Simple Linear Regression
model to the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.
(Include your Python code and output in the code box below.
'''
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

x_mean = np.mean(X)
y_mean = np.mean(Y)

b1 = np.sum((X - x_mean) * (Y - y_mean)) / np.sum((X - x_mean)**2)
b0 = y_mean - b1 * x_mean


Y_pred = b0 + b1 * X


print("Intercept (β0):", b0)
print("Slope (β1):", b1)


plt.scatter(X, Y)
plt.plot(X, Y_pred)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.show()
```

```python
'''
Question 7: Fit a Multiple Linear Regression model on this sample data:
● Area = [1200, 1500, 1800, 2000]
● Rooms = [2, 3, 3, 4]
● Price = [250000, 300000, 320000, 370000]
Check for multicollinearity using VIF and report the results.
'''
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

data = pd.DataFrame({
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
})

X = data[['Area', 'Rooms']]

vif = pd.DataFrame()
vif['Feature'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)
```

```python
'''
Question 8: Implement polynomial regression on the following data:
● X = [1, 2, 3, 4, 5]
3
● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve.
'''

import numpy as np
import matplotlib.pyplot as plt

# Given data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Fit 2nd-degree polynomial
coeff = np.polyfit(X, Y, 2)
model = np.poly1d(coeff)

# Predicted values
X_new = np.linspace(1, 5, 100)
Y_new = model(X_new)

# Print equation
print("Polynomial Equation:", model)

# Plot
plt.scatter(X, Y)
plt.plot(X_new, Y_new)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2nd Degree Polynomial Regression")
plt.show()
```

```python
'''
Question 9: Create a residuals plot for a regression model trained on this data:
● X = [10, 20, 30, 40, 50]
● Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals
'''
import numpy as np
import matplotlib.pyplot as plt


X = np.array([10, 20, 30, 40, 50])
Y = np.array([15, 35, 40, 50, 65])


b1 = np.sum((X - X.mean()) * (Y - Y.mean())) / np.sum((X - X.mean())**2)
b0 = Y.mean() - b1 * X.mean()

Y_pred = b0 + b1 * X

residuals = Y - Y_pred

plt.figure(figsize=(8, 6))
plt.scatter(X, residuals)
plt.axhline(0)
plt.xlabel("X")
plt.ylabel("Residuals")
plt.title("Residuals Plot")
plt.show()
```
