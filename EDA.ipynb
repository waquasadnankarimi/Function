Question-2: Explain overfitting and underfitting in ML. How can you detect and prevent them?

 - Overfitting and underfitting represent two common challenges in machine learning, both impacting a model's ability to generalize to unseen data

 - 1. Overfitting:

     Overfitting occurs when a model learns the training data too well, including its noise and irrelevant details. This results in excellent performance on the training set but poor performance on new, unseen data. It's characterized by high variance and low bias. The model essentially "memorizes" the training examples rather than learning the underlying pattern.

 - 2. Underfitting:

     Underfitting occurs when a model is too simple to capture the underlying patterns in the training data. This leads to poor performance on both the training and test sets. It's characterized by high bias and low variance. The model fails to learn the essential relationships within the data.

   **Detecting Overfitting and Underfitting:**

  - Comparing Training and Validation Performance:
      - Overfitting: High training accuracy/low training error, but significantly lower validation/test accuracy or higher validation/test error.
      - Underfitting: Both training and validation/test accuracy are low, or both training and validation/test errors are high.
  - Learning Curves: Plotting model performance (e.g., error) against the number of training examples or epochs.
     - Overfitting: Training error continues to decrease, while validation error plateaus or increases.
     - Underfitting: Both training and validation errors remain high and relatively flat.
  - Cross-Validation: Evaluating the model's performance on multiple subsets of the data. Large variations in performance across folds can indicate overfitting

  **Preventing Overfitting and Underfitting:**

 - Bias-Variance Tradeoff: The goal is to find a model complexity that balances bias and variance.
   - To reduce high variance (overfitting): Simplify the model, increase training data, use regularization.
   - To reduce high bias (underfitting): Increase model complexity, add relevant features, reduce regularization.
 - Cross-Validation: Techniques like K-fold cross-validation help assess a model's generalization performance more robustly and identify potential overfitting.
 - Regularization Techniques: These methods add a penalty to the loss function based on the complexity of the model, discouraging overly complex models and reducing overfitting.
    - L1 Regularization (Lasso): Adds a penalty proportional to the absolute value of the coefficients, promoting sparsity (some coefficients become zero).
    - L2 Regularization (Ridge): Adds a penalty proportional to the square of the magnitude of the coefficients, shrinking them towards zero.
    - Dropout: In neural networks, randomly deactivates a percentage of neurons during training, preventing co-adaptation and improving generalization.

Question-3:How would you handle missing values in a dataset? Explain at least three methods with examples.

**Handling Missing Values**

  Handling missing values is a crucial step in data preprocessing to ensure data integrity and model performance. Common methods include deletion, simple imputation, and predictive modeling, each suited for different scenarios.

   **Method 1: Deletion**

   Deletion involves simply removing rows or columns with missing values. This method is straightforward but can lead to significant data loss if not used judiciously.
  - How It Works:
     - Listwise deletion (row deletion): The entire record (row) is removed if any single value is missing.
     - Casewise deletion (column deletion): The entire feature (column) is removed if it has too many missing entries.
     - Example: If a dataset of customer information has 1% missing data in the "Age" column across 1000 rows, deleting those 10 rows might be acceptable. However, if 50% of the "Income" column is missing, deleting the entire column might be more appropriate than losing half the dataset rows

**Method 2: Mean/Median Imputation**

  Mean/median imputation replaces missing values with the mean or median of the available values in the same feature (column). This is a simple and fast method that preserves the size of the dataset.
 - How It Works: The average (mean) or the middle value (median) of the non-missing data points for a specific variable is calculated and used to fill the gaps.
   - Example: In a housing price dataset, if some houses are missing their "Square Footage" value, you can fill those gaps with the median square footage of all the other houses in the dataset to mitigate the influence of outliers

 ** Method 3: Predictive Modeling**

  Predictive modeling involves building a machine learning model to predict the missing values based on other variables in the dataset. This sophisticated approach often yields more accurate results than simple imputation.
 - How It Works: One column with missing values acts as the target variable for a separate imputation model, which uses the remaining columns as features to make a prediction.
    - Example: To predict missing "Income" values, you could use a regression model that leverages other features like "Age," "Education Level," and "Occupation" to estimate the most probable income for each individual with a missing value.

Question-4:What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).

  - An imbalanced dataset is one where the distribution of classes within the target variable is unequal. This is a common issue in machine learning, particularly in classification problems, where the number of observations in some classes (majority classes) significantly outweighs the number of observations in other classes (minority classes).
    - For example, in a medical context, a dataset predicting a rare disease might have 99% of samples as "not diseased" and only 1% as "diseased". Standard machine learning algorithms often struggle with this imbalance, as they tend to be biased towards the majority class and perform poorly in predicting the minority class, which is often the class of greatest interest.

**Two techniques to handle an imbalanced dataset are resampling and using synthetic data generation**

  - 1. Technique: Undersampling the Majority Class:

    Undersampling involves reducing the number of samples in the majority class to match the number of samples in the minority class, creating a more balanced dataset.

    - Theoretical:-	The core idea is to balance the class distribution by discarding instances from the abundant class. This changes the prior probability that the learning algorithm sees, theoretically preventing the algorithm from simply predicting the majority class all the time to achieve high accuracy. It aims to reduce the bias towards the majority class
    - Practical:-A simple practical approach is random undersampling. In a dataset with 1,000 "No Disease" examples and 100 "Disease" examples, you would randomly select and discard 900 "No Disease" examples, resulting in a new, balanced dataset of 200 samples. A potential drawback is that you might discard useful information from the majority class during this random removal process.

  - 2. Technique: Synthetic Minority Over-sampling Technique (SMOTE).

    SMOTE is an oversampling technique that addresses class imbalance by creating synthetic examples of the minority class, rather than simply duplicating existing ones
    - Theoretical	SMOTE works by selecting instances that are close to each other in the feature space (using k-nearest neighbors). It draws a line between the data points and generates new, synthetic data points along these lines. This not only balances the dataset but also increases the decision region for the minority class, making it less prone to overfitting than simple oversampling by duplication.
    - Practical	Using a library like imblearn in Python, you can apply SMOTE to a dataset. The algorithm identifies minority samples, finds their nearest neighbors, and creates new samples at interpolated points. This effectively increases the representation of the rare class without losing the original majority data, leading to a robust, balanced training dataset.

Question-5: Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.

Feature scaling is important in machine learning because many algorithms are sensitive to the magnitude and units of input features. By scaling, you prevent features with larger values from dominating the learning process, which ensures the model converges more efficiently and performs better.

**Min-Max Scaling vs. Standardization**

The two most common methods, Min-Max scaling and standardization, transform data differently and are suited for different situations.
Min-Max Scaling
  - Process: This technique transforms features by scaling each value to a fixed range, typically between 0 and 1. The formula uses the minimum and maximum values of the dataset to achieve this.
  - Best For: It's ideal for algorithms that expect input features in a specific bounded range (e.g., neural networks) or when the original data distribution is not Gaussian.

**Standardization (Z-Score Normalization)**

 - Process: Standardization transforms the data to have a mean of zero and a standard deviation of one (the z-score). It centers the data distribution around the mean, regardless of the original range.
 - Best For: It is particularly useful for algorithms that assume a Gaussian distribution in the input features, such as linear regression, logistic regression, and Support Vector Machines, as it is less affected by outliers than Min-Max scaling.

In essence, Min-Max scaling bounds the data within a specific range, while standardization rescales the data to a standard distribution with a mean of 0 and standard deviation of 1. The choice depends on the specific algorithm and the nature of the data distribution

Question-6: Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?

 - One-Hot Encoding creates a new binary column for each category, which is better for nominal data (where order doesn't matter), like "red," "green," "blue," to prevent the model from assuming a false hierarchy

**Label Encoding**

 - When to use: For ordinal data (e.g., education levels: High School, Bachelor's, Master's) where the numerical values have a natural order. It's also efficient for nominal data with a large number of unique values, especially with tree-based models.
 - How it works: Replaces each category with a unique integer. For example, {'low': 0, 'medium': 1, 'high': 2}.
 - Pros:
   - Computationally efficient and uses less memory.
   - Simple to implement.
 - Cons:
   - Introduces an artificial ordinal relationship in nominal data, which can mislead many algorithms (especially linear and distance-based ones).
   - The magnitude of the integer can be misinterpreted by the model (e.g., that "high" (2) is twice "medium" (1)).

**One-Hot Encoding**
 - When to use: For nominal data (e.g., colors, city names) with a limited number of unique categories to avoid introducing a false ordinal relationship.
 -How it works: Creates a new binary (0 or 1) column for each unique category. For example, "red" might become [1, 0, 0], "green" [0, 1, 0], and "blue" [0, 0, 1].
 - Pros:
   - Prevents algorithms from assuming any order or hierarchy among categories.
   - Works well with linear and distance-based models.
 - Cons:
   - Can lead to a very high number of features (high dimensionality) if the category has many unique values, which can be computationally expensive and lead to the "curse of dimensionality".
   - Can create a sparse matrix, as most values will be zero.





```python
'''
Question 7: Google Play Store Dataset
a). Analyze the relationship between app categories and ratings.
Which categories have the highest/lowest average ratings,
and what could be the possible reasons?
Dataset: https://github.com/MasteriNeuron/datasets.git
(Include your Python code and output in the code box below.)

'''
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset â€” adjust path/URL as needed
# Example: if the dataset CSV is named 'apps.csv'
df = pd.read_csv('/googleplaystore.csv')

# Inspect top rows
df.head()

# Clean data: remove rows with missing Rating or missing Category
df_clean = df.dropna(subset=['Category', 'Rating'])

# Ensure Rating is numeric (convert if necessary)
df_clean['Rating'] = pd.to_numeric(df_clean['Rating'], errors='coerce')

# Drop rows where conversion failed (if any)
df_clean = df_clean.dropna(subset=['Rating'])

# Compute average rating per category
cat_rating = df_clean.groupby('Category')['Rating'].mean().sort_values(ascending=False)

print("Average rating by category (descending):")
print(cat_rating)

# Also compute count of apps per category (so you know how many contributed)
cat_count = df_clean['Category'].value_counts()

print("\nNumber of apps per category:")
print(cat_count)

# Combine into a DataFrame for clarity
summary = pd.DataFrame({'count': cat_count, 'avg_rating': cat_rating})
summary = summary.dropna().sort_values(by='avg_rating', ascending=False)

print("\nSummary (categories with enough apps):")
print(summary.head(10))  # top 10 by rating
print(summary.tail(10))  # bottom 10 by rating

# (Optional) Visualize top and bottom 10 categories by average rating
plt.figure(figsize=(12,6))
summary['avg_rating'].head(10).plot(kind='bar', color='green')
plt.title('Top 10 Categories by Average Rating')
plt.ylabel('Average Rating')
plt.show()

plt.figure(figsize=(12,6))
summary['avg_rating'].tail(10).plot(kind='bar', color='red')
plt.title('Bottom 10 Categories by Average Rating')
plt.ylabel('Average Rating')
plt.show()

```

```python
'''
Question 9: Flight Price Prediction Dataset
a) How do flight prices vary with the days left until departure? Identify any exponential price
surges and recommend the best booking window.
b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
consistently cheaper/premium, and why?
Dataset: https://github.com/MasteriNeuron/datasets.git
'''
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset â€” adjust filename/path accordingly
df = pd.read_csv('/flight_price.csv')

df.head()

# Some basic cleaning (if needed)
df = df.dropna(subset=['days_left', 'price', 'airline', 'source_city', 'destination_city'])

# Convert data types if needed
df['days_left'] = df['days_left'].astype(int)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df = df.dropna(subset=['price'])

# --- PART (a): Price vs Days Left Until Departure ---
plt.figure(figsize=(10,6))
sns.scatterplot(x='days_left', y='price', data=df, alpha=0.3)
plt.title('Flight Price vs Days Left Until Departure')
plt.xlabel('Days Left Until Departure')
plt.ylabel('Ticket Price')
plt.show()

# Average price by days_left (to smooth noise)
avg_price = df.groupby('days_left')['price'].mean()
plt.figure(figsize=(10,6))
avg_price.plot()
plt.title('Average Ticket Price vs Days Left')
plt.xlabel('Days Left')
plt.ylabel('Average Price')
plt.show()

print("Some sample of average price by days left:")
print(avg_price.head(20))
print(avg_price.tail(20))

route_df = df[(df['source_city']=='Delhi') & (df['destination_city']=='Mumbai')]

plt.figure(figsize=(12,6))
sns.boxplot(x='airline', y='price', data=route_df)
plt.title('Price distribution by Airline for route Delhi â†’ Mumbai')
plt.xticks(rotation=45)
plt.ylabel('Ticket Price')
plt.xlabel('Airline')
plt.show()

avg_airline_price = route_df.groupby('airline')['price'].mean().sort_values()
print("Average ticket price by airline on Delhiâ€“Mumbai route:")
print(avg_airline_price)
```

```python
'''
Question 10: HR Analytics Dataset
a). What factors most strongly correlate with employee attrition?
Use visualizations to show key
drivers (e.g., satisfaction, overtime, salary).
b). Are employees with more projects more likely to leave?
Dataset: hr_analytics
'''

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset â€” adjust filename/path accordingly
df = pd.read_csv('/hr_analytics.csv')

df.head()

df['Attrition_Flag'] = df['Attrition'].map({'Yes':1, 'No':0})

correlations = df.corr()['Attrition_Flag'].sort_values(ascending=False)
print("Correlation with Attrition:")
print(correlations)

# Visualize top 5 factors
top_factors = correlations[1:6]  # skip Attrition_Flag itself
top_factors.plot(kind='bar', title='Top Factors Correlated with Attrition')
plt.show()

projects_attrition = df.groupby('NumProjects')['Attrition_Flag'].mean()
print("\nAverage Attrition Rate by Number of Projects:")
print(projects_attrition)

# Plot
projects_attrition.plot(marker='o', title='Attrition vs Number of Projects')
plt.xlabel('Number of Projects')
plt.ylabel('Average Attrition Rate')
plt.grid(True)
plt.show()



```
